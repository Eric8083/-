В некотором царстве, В некотором государстве, жил принц, счастливый и богатый. Но в какой-то момент его отец, Великий Велик Великович Величайший скончался, и пошел вопрос об разделении его богатсв по родственникам. Его украшения достались его 2 дочерям. Каждой дано по 3 украшения, но оказалось, что украшений было 5, и одни повторялись. Так что некоторые украшения были приписанны двум дочерям одновременно. Требуется найти повторения и их исправить(Если есть пара одинаковых, одно никому не давать), а затем распределить украшен
равномерно.

---

Формат Ввода:
Название каждого украшения через запятую и пробел(Гарантируется, что будут повторяющиеся)



---

Формат Вывода:

Украшения, доставшися первой, а на новой строчке - второй



---

Пример:

Ввод:

Золотая цепь, Алмазная цепь, рубиновое кольцо, Золотая цепь, эмеральдные серги

Вывод:

1. Золотая цепь, эмеральдные серги
2. Алмазная цепь, рубиновое кольцо

Для сбора лицейского меню опросили N количество школьников, какие вторые блюда они хотели бы видеть в столовой. Опрошенным давали +5% к их баллу в поступление в лицей. Напишите программу, которая после ввода данных Выведет блюда, которые предложены в меню, а затем учеников, получивших бонус к поступлению.


---

Формат Ввода:

N, затем N Количество строк, на каждой ученик и его предложение в меню.


---

Формат вывода:

1.Cписок блюд
2.Список учеников


---

Пример:

Ввод:

4

Вася Пюре

Петя Макароны

Ваня Мясо

Даша Рыба

Вывод:

Пюре Макароны Мясо Рыба

Вася, Петя, Ваня, Даша

```python
stud_food=dict()
N = int(input())
for i in range(N):
   phrase =input().split()
   student = phrase[0]
   food = ' '.join(phrase[1:])
   stud_food.setdefault(food, student)
unique_dishes = list(stud_food.keys()) #Как дополнительный шаг убираю повторы, чтобы результат был чище
name = list(stud_food.values())
print("Список блюд: ",unique_dishes)
print("Список учеников: ", name)
```
